# Imports

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

# Networkx test

```python
## Creating Nodes (I am not sure if it connects them)
gr = nx.Graph()
node_list = []
for i in range(4):
    node_list.append(i)
gr.add_nodes_from(node_list)
```

```python
### Creating edges basically creates the nodes while adding an edge
### This is creating a mesh network where each vertex or node is connected to every other node/vertex (I believe this is how DSRC works)
gr1 = nx.Graph()
for i in range(5):
    for j in range(5):
        if j != i:
            gr.add_edge(i,j)
            # print(f'Added edge {i,j}')
print(gr1.number_of_nodes())
print(gr1.number_of_edges())
```

```python
#using the complete_graph feature (connecting each node to every other node)
gr2 = nx.complete_graph(5)
```

```python
# nx.draw(gr, with_labels = True, font_weight = 'bold')
nx.draw_shell(gr2, with_labels = True, font_weight = 'bold')
plt.show()
```
